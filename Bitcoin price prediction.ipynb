```python
#Program to predict bitcoin price for the next 30 days(the first project)
#importing libraries
import numpy as np 
import pandas as pd
```

```python
#loading the dataset
df = pd.read_csv('market-price')
```

```python
#exploring first 7 days n last 7 days of market price
print(df.head(7))
print(df.tail(7))
```

```python
df.drop(['Timestamp'],1,inplace=True)
```

```python
df
```

```python
# A Variable for predicting 'n' days out into the future
prediction_days = 30

# Create another column shifted 'n' units up
df['prediction']=df[['market-price']].shift(-prediction_days) #we created new column by minusing first 30 days
```

```python
df.head()
```

```python
#to view last seven days of new dataset
df.tail(7)      #so,that last 30 days price will be removed
```

```python
# Creating the independent data-set

# converting the dataframe into numpy array & removing prediction column

X=np.array(df.drop(['prediction'],axis=1))
```

```python
X=X[:len(df)-prediction_days]
```

```python
#CREATE THE DEPENDENT DATA SET (y) 
# Convert the dataframe to a numpy array (All of the values including the NaN's)
y = np.array(df['prediction'])  
# Get all of the y values except the last 'n' rows 
y = y[:-prediction_days] 
print(y)
```

```python
# Split the data into 80% training and 20% testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Set prediction_days_array equal to the last 30 rows of the original data set from the price column
prediction_days_array = np.array(df.drop(['prediction'],1))[-prediction_days:] #is ignoring last 30 values 4 drop
print(prediction_days_array)
```

```python
from sklearn.svm import SVR
# Create and train the Support Vector Machine 
svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.00001)#here using c to control error & gamma-how far influence of a single training example reaches.
svr_rbf.fit(x_train, y_train) #Train the model
```

```python
# Testing Model: Score returns the accuracy of the prediction. 
# The best possible score is 1.0
svr_rbf_confidence = svr_rbf.score(x_test, y_test)
print("svr_rbf accuracy: ", svr_rbf_confidence)
```

```python
# Print the predicted value
svm_prediction = svr_rbf.predict(x_test)
print(svm_prediction)

print()

#Print the actual values
print(y_test)
```

```python
# Print the model predictions for the next 'n=30' days
svm_prediction = svr_rbf.predict(prediction_days_array)
print(svm_prediction)
```
